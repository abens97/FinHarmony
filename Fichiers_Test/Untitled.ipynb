```python
import pandas as pd
from IPython.display import display, HTML
```

```python
%%time
nom = "Safran_X"
dataframe_list = pd.read_html(io = nom +'.xhtml', )#match='balance'
dataframe_list_filtered = [i for i in dataframe_list if len(i) > 2]
```

```python
%%time
# Create a Pandas Excel writer using XlsxWriter as the engine.
writer = pd.ExcelWriter(nom + '.xlsx', engine='xlsxwriter')

for i in range (len(dataframe_list)) :
    print(i+1)
    dataframe_list[i].to_excel(writer, sheet_name=str(i+1))

# Close the Pandas Excel writer and output the Excel file.
writer.save()
```
